# Header conflict protein.fasta and CDS.fasta

 - While CDS.fasta files have sequential headers e.g. main1, main2 etc, the protein.fasta header produced by gffread is not sequential.
 - To ensure that the protein prediction can be linked to CDS ID, use maker scripts to change header within gff then remake protein.fasta and CDS.fasta

http://weatherby.genetics.utah.edu/MAKER/wiki/index.php/MAKER_Tutorial_for_WGS_Assembly_and_Annotation_Winter_School_2018 #rename gene models and other data.

The script maker_map_ids produces a map file of original header paired to sequential headers.
The script map_gff_ids takes the map file and renames the gff file.

The scripts were copied from /bioinf_resources/maker/bin
