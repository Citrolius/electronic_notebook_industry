###### Os06g0164900

Os06g0164900. p-value 5.29E-08. LogP 7.276544328. Knetscore 8.9

 - Occurs on chromosome 6 and influences germination rate
 - low knetscore possibly due to the paper discussing primarily the link between germination and immunity.
 

 - OS06G0164900 encodes for OS06T0164900-00 protein which bears sequence similarity with LECRK1, G-type lectin S-receptor-like serine/threonine-protein kinase.
 
 - The study desginates OS06T0164900-00 protein as OSlecRK
 
 - Publication PMID 24033867. Knock down of OslecRK inhibits -amylase genes expression. This leads to decreased germination rate. However, it also suppresses key immunity genes.
 
 - The study also questions if the successful survival of plants during germination is caused by strong vigor or if vigor is only a correlation. I.e. strong vigour leads to bettter fungal and bacterial resistance and it is actually strong immune system that results in increased survival during germination. 
 
 - Could it be much more feasible to boost the immunity of rice instead of merely trying to increase ESVs?

#### False positive OS06G0156700

OS06G0156700 does indeed have publications with content about root growth but the publication highlights that although OS06G0156700's cDNA could control tissue-specific expression during germination (high expression of OS06G0156700) it doesn't influence root growth, merely morphogenesis.


This is despite a higher knetscore than OS06G0164900 at 21.91

###### Os01g0607900

The condensed network view shows that OS01G0607900 is orthologous to AT2G26730 in Arabidopsis which in turn is physically intracting with 2 different genes. Of particular significance is NHL6 which according to the publications plays a key role in abiotic stress induced ABA (abscisic acid hormone) signalling pathway. In particular if the ortholog in Arabidopsis is down regulated, the seedling is less sensitive to ABA which stimulates germination and early seedling growth afterwards. (PMID 26849212 )

The other gene which is physically regulated by AT2G26730 was involved in a study suggesting the receptor like kinase encoded by a similar gene in sweet potatoes it may influence root thickening. However, this is less relevant as the study was conducted on somatic stem tissue instead of on seedlings.
