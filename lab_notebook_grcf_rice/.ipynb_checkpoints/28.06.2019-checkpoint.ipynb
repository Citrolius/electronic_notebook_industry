###### From yesterday's research, the hdf5 files from AraGWAS contains complex groups instead of mere datasets.


  - hdf5 files are hierachical data formats as if the file itself is a parent directory.
 
  - All content are stored as binary. 
  
  - They are useful for storing large datasets and access  a specific subset easily because it utilises Python numpy.
  
  - Metadata can also be added to every element in structure.
  
  - h5py library does not read the entire file into ram memory, only the subset.
